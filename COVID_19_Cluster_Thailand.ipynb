```
import pandas as pd
import altair as alt
import numpy as np
```

```
url='https://data.go.th/dataset/8a956917-436d-4afd-a2d4-59e4dd8e906e/resource/93f74e67-6f76-4b25-8f5d-b485083100b6/download/covid-19-daily.xlsx' #read data of all patients
df=pd.read_excel(url,na_values=['ไม่ระบุ','-','ไม่ทราบ'])
df["age"]=["วัยเด็ก" if i <= 10 else "วัยรุ่น" if i<=20 else "วัยทำงาน" if i<=60 else "ผู้สูงอายุ" for i in df["age"]] 
df['count_value']=[1 for i in range(len(df['no']))]
df[df["Announce Date"] =="2020-03-12"] #look what is start index for 2020-03-12
```

```
df[df["Announce Date"] =="2020-04-03"] #look what is end index for 2020-04-03
```

```
df_new=df.iloc[59:1978]
df_new #select only 2020-03-12 - 2020-04-03 
```

```
x=df_new.groupby("Announce Date").count()["no"]
count_all=np.array(x)
```

```
url_2 ="https://github.com/chinnapaht/COVID-19-Visualization/blob/master/Covid-19-Clusters-Thailand-acc%20.csv" # read cluster data in thailand
df_2=pd.read_html(url_2)[0]
df_2
```

```
df_3=df_2.groupby("สาเหตุ/สถานที่").sum()
x=np.array(df_3["จำนวนผู้ป่วยในแต่ละวัน"])
df_3["percent"]= np.around((x/sum(count_all))*100)
df_3["สาเหตุ/สถานที่"]=df_3.index
df_3
```

```
selection = alt.selection_multi(fields=['สาเหตุ/สถานที่'], bind='legend')

chart1=alt.Chart(df_2).mark_area().encode(
    x="วัน:O",
    y='จำนวนผู้ป่วยในแต่ละวัน',
    color=alt.Color('สาเหตุ/สถานที่:N',scale=alt.Scale(scheme='tableau20')),tooltip= ["วัน:O",'จำนวนผู้ป่วยในแต่ละวัน','สาเหตุ/สถานที่'],
     opacity=alt.condition(selection, alt.value(1), alt.value(0.2))).add_selection(
    selection
).properties(
    title='จำนวนผู้ติดเชื้อ COVID-19 ระหว่างวันที่ 12 มีนา 62 - 3เมษา 62 แบ่งตามสาเหตุและสถานที่'
)
    

chart2 = alt.Chart(df_3).mark_bar().encode(
    y="สาเหตุ/สถานที่:O",
    x= alt.X('จำนวนผู้ป่วยในแต่ละวัน',title = "จำนวนผู้ป่วย"),
    color=alt.Color('สาเหตุ/สถานที่:N', scale=alt.Scale(scheme='tableau20')),tooltip= ['จำนวนผู้ป่วยในแต่ละวัน','สาเหตุ/สถานที่','percent'],
    opacity=alt.condition(selection, alt.value(1), alt.value(0.2))
    ).add_selection(
    selection
).properties(
    title='จำนวนผู้ติดเชื้อ COVID-19 ระหว่างวันที่ 12 มีนา 62 - 3เมษา 62 แบ่งตามสาเหตุและสถานที่'
)

chart3=alt.Chart(df_3).mark_bar().encode(
    y="สาเหตุ/สถานที่:O",
    x=alt.X('จำนวนผู้ป่วยในแต่ละวัน',title = "จำนวนผู้ป่วยเทียบเป็นเปอร์เซ็นต่อผู้ป่วยทั้งหมด"),
    color=alt.Color('สาเหตุ/สถานที่:N', scale=alt.Scale(scheme='tableau20')),tooltip= ['จำนวนผู้ป่วยในแต่ละวัน','สาเหตุ/สถานที่','percent'],
    opacity=alt.condition(selection, alt.value(1), alt.value(0.2))
    ).add_selection(
    selection
).properties(
    title='จำนวนผู้ติดเชื้อ COVID-19 แบ่งตามสาเหตุและสถานที่เทียบเป็นเปอร์เซ็นของผู้ที่ได้รับเชื้อระหว่าวันที่  12 มีนา 62 - 3เมษา 62'
)

chart1|(chart2&chart3)
```
