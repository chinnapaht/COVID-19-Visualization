```
url='https://data.go.th/dataset/8a956917-436d-4afd-a2d4-59e4dd8e906e/resource/93f74e67-6f76-4b25-8f5d-b485083100b6/download/covid-19-daily.xlsx'
df=pd.read_excel(url,na_values=['ไม่ระบุ','-','ไม่ทราบ'])
df["age"]=["วัยเด็ก" if i <= 10 else "วัยรุ่น" if i<=20 else "วัยทำงาน" if i<=50 else "ผู้สูงอายุ" for i in df["age"]] 
df['count_value']=[1 for i in range(len(df['no']))]
df
```

```
chart1=alt.Chart(df).mark_bar().encode(
    x='age',
    y='sum(count_value)',
    color='sex'
)
chart2=alt.Chart(df_2).mark_bar().encode(
    x="patient",
    y='สาเหตุหรือสถานที่ที่ติดเชื้อ')
chart1|chart2
```
